```python
df=df.drop(['Unnamed: 0'],axis=1)
```

```python
# Since the gps coordinates are available, the coordinates are used to
# calculate the distance for tourist destination or "anchors" such as
# jordaan neighborhood (a famous neighborhood in amestrdam), Van Gogh museum, Weesperbuurt en Plantage(a famous neighborhood in amestrdam)	, and Vondelpark
import geopy.distance
# Coordinates are pulled from Google Maps
jordaan_coords=(52.37, 4.88)
VanGogh_coords = (52.35, 4.88)
Weesperbuurt_en_Plantage_coords = (52.36, 4.90)
Vondelpark_coords = (52.35, 4.86)
df['dist_jorda'] = 0.0
df['dist_vg'] = 0.0
df['dist_wep'] = 0.0
df['dist_vp'] = 0.0
for index, row in df.iterrows():
    #print(row['latitude'])
    print(index)
    df['dist_jorda'][index] = geopy.distance.vincenty((row['latitude'],row['longitude']),jordaan_coords).miles
    df['dist_vg'][index] = geopy.distance.vincenty((row['latitude'],row['longitude']),VanGogh_coords).miles
    df['dist_wep'][index] = geopy.distance.vincenty((row['latitude'],row['longitude']),Weesperbuurt_en_Plantage_coords).miles
    df['dist_vp'][index] = geopy.distance.vincenty((row['latitude'],row['longitude']),Vondelpark_coords).miles
```

```python
df = df.drop(['latitude','longitude'], axis = 1)
print ('Distance to tourist destination variables were added')
print
```

```python
# The amenities variable appears to be a list of amenities, which
# needs to be split out into separate amenities using dummy variables
df['amenities_split'] = df['amenities'].apply(lambda x: x.replace("  ", '').replace("_", ' ').replace("]", '').replace("* Use 'Liebherr' ", '').replace('* I wash the dishesrefrigerator', '').replace('. ', '').replace('??? ', '').replace('77 ', '').replace('75 ', '').replace('70 ', '').replace('65 ', '').replace('60 ', '').replace('57 ', '').replace('53 ', '').replace('52 ', '').replace('50 ', '').replace('49 ', '').replace('48 ', '').replace('47 ', '').replace('45 ', '').replace('44 ', '').replace('43 ', '').replace('42 ', '').replace('40 ', '').replace('38 ', '').replace('36 ', '').replace('32 ', '').replace('.  ', '').replace('- ', '').replace('[', '').replace('[', '').replace(']', ',').replace('{', '').replace('\\', '').replace('55\\', '').replace('}', ',').replace('"', '').split(','))
df = pd.concat((df.drop(['amenities','amenities_split'],axis=1), pd.get_dummies(df['amenities_split'].apply(pd.Series).stack(),prefix='a', prefix_sep='_').sum(level=0)),axis = 1)
print ('Dummy variables were create for every single amenity with the Amentities variable')
print('')
```

# Detecting 'price' Outlier Using Quantile Percentage

```python
# We can consider values above 95% and 5% Quantile as outliers.We will use same criteria to sort out the outliers
max_thresold = df['price'].quantile(0.95)
max_thresold
```

## Data Above Max Threshold¬∂


```python
df[df['price']>max_thresold].head(2)
```

```python
# falls above the maximum threshold of 0.95 Quantile.
```

## Minmum Threshold¬∂


```python
min_thresold = df['price'].quantile(0.05)
min_thresold
```

```python
# Data Below Minimum Threshold¬∂
```

```python
df[df['price']<min_thresold].head(2)
```

```python
# Remove outliers for price column
df=df[(df['price']<max_thresold) & (df['price']>min_thresold)]
df
```

# lets remove the outliers for the rest of our numerical columns

```python
def cap_data(df):
    for col in df.select_dtypes(include=numerics):
        print("capping the ",col)
        if (((df[col].dtype)=='float64') | ((df[col].dtype)=='int64')):
            percentiles = df[col].quantile([0.01,0.99]).values
            df[col][df[col] <= percentiles[0]] = percentiles[0]
            df[col][df[col] >= percentiles[1]] = percentiles[1]
        else:
            df[col]=df[col]
    return df

df=cap_data(df)
```
